##  Normal distribution
The following represent the continuos gauss distribution

$$P(X) = \frac{1}{\sigma \sqrt{2 \pi}} \exp{\left[-\frac{1}{2}\left(\frac{X-\mu}{\sigma} \right)^2 \right]} \space{(1)}$$

* $\mu$: mean of the distribution
* $\sigma$: standard desviations of the distribution

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```
def gaussian(x,mu, sigma):          #we define the gauss function
    return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-0.5 * pow((x-mu)/sigma,2))
```

```
x = np.arange(-4,4,0.1)            #we create a list of data to apply the fuction
y = gaussian(x, 1.0, 1.0)          #we create y values
plt.plot(x,y)                      # we graph the distribution
```

The fast way to do normal distribution is scipy library

```
dist = norm(0,1)                              #we use the library to do the same
x =np.arange(-4,4,0.1)                        #the same data for x values to apply to the function
y = [dist.pdf(value) for value in x]       #pdf = probability density function = distribution  
plt.plot(x,y)
```

The accumultive is the integer from (1) scypi we used cdf for cumulative density function easy to do like this: 

```
dist = norm(0,1)
x = np.arange(-4,4,0.1)
y = [dist.cdf(values) for values in x]            #we use cdf=cumulative density function
plt.plot(x,y)                                     #cumulative distribution plot associate to gauss function
```
